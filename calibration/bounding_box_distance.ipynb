```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import leastsq
```

```python
x = np.array([2.68, 5.30072, 10.1537, 15.1029, 20.07729, 25.06188, 30.05158, 35.04422]) # cone_dist
y = np.array([0.17222, 0.10277, 0.06111, 0.04166, 0.03055, 0.0222, 0.01944, 0.0166]) # rel_cone_img_hight
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(x, y, color='r')
ax.set_xlabel('Cone distance')
ax.set_ylabel('relative hight in image')
ax.set_title('scatter plot')
plt.show()
```

```python
initial_guess = [1]

def function_to_fit(x, p):
    return 1/(1+p[0]*x)

def residuals(p,y,x):
    return y - function_to_fit(x,p)

cnsts = leastsq(
    residuals, 
    initial_guess, 
    args=(y, x)
)[0]

print("New coefficients: ", cnsts)

fig, ax = plt.subplots()
ax.plot(x, y, 'o')

xi = np.arange(0,40,1)
ax.plot(xi, [function_to_fit(x, cnsts) for x in xi])

plt.show()
```
